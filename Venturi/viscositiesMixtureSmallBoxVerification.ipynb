```python
fig1 = plt.gcf()
plt.plot(data.YCH4, data.DavidsonSU2, color='cornflowerblue', linewidth=2)
plt.plot(data.YCH4, data.DavidsonStandalone, '--', color='darksalmon', linewidth=2)
plt.plot(data.YCH4, data.WilkeSU2, color='darkgreen', linewidth=2)
plt.plot(data.YCH4, data.WilkeStandalone, '--', color='limegreen', linewidth=2)
plt.xlabel('YCH4 [-]', fontsize = 14)
plt.ylabel('$\mu_{mix}$ [kg/(m$\cdot$s)]', fontsize = 14)
plt.xlim(0,1)
plt.ylim(1e-5,2e-5)
plt.legend(('Davidson SU2', 'Davidson Standalone', 'Wilke SU2', 'Wilke Standalone'))
#plt.text(0.85,0.3e-5,'T100K', fontweight='bold', bbox={'facecolor': 'white', 'alpha': 1, 'pad': 2.5})
#plt.text(0.85,1.225e-5,'T323K', fontweight='bold', bbox={'facecolor': 'white', 'alpha': 1, 'pad': 2.5})
#plt.text(0.85,1.4e-5,'T373K', fontweight='bold', bbox={'facecolor': 'white', 'alpha': 1, 'pad': 2.5})
#plt.text(0.85,1.575e-5,'T423K', fontweight='bold', bbox={'facecolor': 'white', 'alpha': 1, 'pad': 2.5})
#plt.text(0.85,3.2e-5,'T1100K', fontweight='bold', bbox={'facecolor': 'white', 'alpha': 1, 'pad': 2.5})
plt.grid(True, which='major')
plt.minorticks_on()
plt.grid(b=True, which='minor', linestyle='-', alpha=0.2)
plt.figure(figsize=(200,100))
fig1.savefig('viscositiesMixtureSmallBoxVerification.jpg', bbox_inches="tight", dpi=200)
```
