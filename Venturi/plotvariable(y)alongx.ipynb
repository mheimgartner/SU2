```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from glob import glob
filenamesFluent = sorted(glob('variables_Fluent_filenr_*.csv'))
filenamesSU2 = sorted(glob('variables_SU2_filenr_*.csv'))
dataFluent = [pd.read_csv(file) for file in filenamesFluent]
dataSU2 = [pd.read_csv(file) for file in filenamesSU2]
for k in range(len(dataFluent)):
    dataFluent[k] = dataFluent[k].dropna()
for l in range(len(dataSU2)):
    dataSU2[l] = dataSU2[l].dropna()
header = [0.075, 0.1, 0.125, 0.15, 0.175, 0.2, 0.233567841]
dataToPrint = [1, 0, 0, 1, 0, 0, 1]
```

```python
fig1 = plt.gcf()
for i in range(len(dataFluent)):
    if dataToPrint[i] == True: 
        plt.plot(dataSU2[i].arc_length, dataSU2[i].Passive_Scalar, linewidth=2, label='SU2 | x = %0.3f' % header[i])
        plt.plot(dataFluent[i].arc_length, dataFluent[i].ch4, '--', linewidth=2, label='Fluent | x = %0.3f' % header[i])
plt.xlabel('y [m]', fontsize = 14)
plt.ylabel('YCH4 [-]', fontsize = 14)
plt.xlim(0,0.0152)
plt.ylim(0,1)
plt.legend()
plt.grid(True, which='major')
plt.minorticks_on()
plt.grid(b=True, which='minor', linestyle='-', alpha=0.2)
plt.figure(figsize=(50,25))
fig1.savefig('YCH4(y)alongx.jpg', bbox_inches="tight", dpi=100)
```

```python
fig2 = plt.gcf()
for i in range(len(dataFluent)):
    if dataToPrint[i] == True:
        plt.plot(dataSU2[i].arc_length, dataSU2[i].Velocity_X, linewidth=2, label='SU2 | x = %0.3f' % header[i])
        plt.plot(dataFluent[i].arc_length, dataFluent[i].axial_velocity, '--', linewidth=2, label='Fluent | x = %0.3f' % header[i])
plt.xlabel('y [m]', fontsize = 14)
plt.ylabel('Axial velocity [m/s]', fontsize = 14)
plt.xlim(0,0.0152)
plt.ylim(0,70)
plt.legend()
plt.grid(True, which='major')
plt.minorticks_on()
plt.grid(b=True, which='minor', linestyle='-', alpha=0.2)
plt.figure(figsize=(50,25))
fig2.savefig('AxialVelocity(y)alongx.jpg', bbox_inches="tight", dpi=100)
```
