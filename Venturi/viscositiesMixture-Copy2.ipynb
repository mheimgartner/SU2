```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
dataBzowski = pd.read_csv('BzowskiData.csv') 
dataSU2BzowskiFull = pd.read_csv('viscosityMixtureBzowskiFullSutherland.csv') 
```

```python
fig1 = plt.gcf()
plt.plot(dataSU2BzowskiFull.Temperature, dataSU2BzowskiFull.viscosityWilke, color='cornflowerblue', linewidth=2)
plt.plot(dataSU2BzowskiFull.Temperature, dataSU2BzowskiFull.viscosityDavidson, '--', color='darksalmon', linewidth=2)
plt.plot(dataBzowski.Temperature, dataBzowski.Viscosity, 'ko',markersize=4)
plt.xlabel('T [K]', fontsize = 14)
plt.ylabel('$\mu_{mix}$ [kg/(m$\cdot$s)]', fontsize = 14)
plt.xlim(100,3400)
plt.ylim(1e-5,9e-5)
plt.legend(('Wilke', 'Davidson', 'Bzowski et al. (1990)'))
#plt.text(1700,5.5e-5,'3.4%', fontweight='bold', color='darksalmon')
#plt.text(2200,6.25e-5,'6.2%', fontweight='bold', color='darksalmon')
#plt.text(2700,7.15e-5,'9.0%', fontweight='bold', color='darksalmon')
#plt.text(3050,8.15e-5,'11.8%', fontweight='bold', color='darksalmon')
#plt.text(1700,6.05e-5,'2.4%', fontweight='bold', color='cornflowerblue')
#plt.text(2200,6.65e-5,'5.1%', fontweight='bold', color='cornflowerblue')
#plt.text(2700,7.65e-5,'7.9%', fontweight='bold', color='cornflowerblue')
#plt.text(3050,8.65e-5,'10.7%', fontweight='bold', color='cornflowerblue')
plt.grid(True, which='major')
plt.minorticks_on()
plt.grid(b=True, which='minor', linestyle='-', alpha=0.2)
plt.figure(figsize=(200,100))
fig1.savefig('viscositiesMixtureVariousTBzowskiFullSutherland.jpg', bbox_inches="tight", dpi=200)
```
